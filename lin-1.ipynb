Algorithm: Linear Regression
Type: Supervised Learning
Reason for choice:
Linear Regression is simple, interpretable, and ideal for understanding how features affect predictions.

```python
x = df.drop(["Country","Life expectancy","thinness  1-19 years","thinness 5-9 years","Income composition of resources"],axis =1)
y = df["Life expectancy"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("R2 Score:", r2)
```
